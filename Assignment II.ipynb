The students will use dataset “crochet.xlsx” and “patada.xlsx”. These were read with a boxing bag with a
symmetric arrangement of 9 3‐axis accelerometers. Each entry consists of 10 lectures per accelerometer
axis (270 real components) taken in 2 ms after an impact, plus an impact label and a power measurement.

- Classification. Students will design SVM models for impact classification. After obtaining the models, students are expected to evaluate them according to diverse quality metrics and compare them with the Null Model (Baseline) and to each other. The design goal is, for a similar performance, minimize  the number of features and the number of support vectors.


- Clustering. Students are expected to apply hierarchical and k‐means clustering algorithms to the datasets. For the later, the value of k may be guessed applying known heuristics.
